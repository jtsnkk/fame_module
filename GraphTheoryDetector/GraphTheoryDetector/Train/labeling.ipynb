# Data Labeling

```python
import pandas as pd
```

```python
mal = pd.read_csv(r'Malware_FCG_Feature.csv',header=None)
benign = pd.read_csv(r'Benignware_FCG_Feature.csv',header=None,error_bad_lines=False)
fam = pd.read_csv('dataset.csv')
```

```python
mal.rename(columns={0:'filename'},inplace=True)
benign.rename(columns={0:'filename'},inplace=True)
```

```python
len(mal),len(benign)
```

```python
data = benign.append(mal)
del mal
del benign
data
```

```python
data = data.merge(fam[['filename','label','CPU Architecture']])
data
```

```python
# get all the label and generate a table for encoding
LabelEncoding = dict()
for i,label in enumerate(list(data.label.value_counts().index)):
    LabelEncoding[label] = i
LabelEncoding
```

```python
# get all the CPU and generate a table for encoding
CPUEncoding = dict()
for i,CPU in enumerate(list(data['CPU Architecture'].value_counts().index)):
    CPUEncoding[CPU] = i
CPUEncoding
```

```python
data.replace(LabelEncoding,inplace=True)
data.replace(CPUEncoding,inplace=True)
data
```

```python
data.to_csv('GraphTheoryDataset.csv')
```
